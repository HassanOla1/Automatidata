#### Understand the data - Investigate the variables

This block of code converts the date columns in the dataset from string format to datetime format.
* This conversion is necessary to perform date-based operations and analysis

```python
# Convert date columns to datetime format
automatidata['tpep_pickup_datetime'] = pd.to_datetime(automatidata['tpep_pickup_datetime'])
automatidata['tpep_dropoff_datetime'] = pd.to_datetime(automatidata['tpep_dropoff_datetime'])
```

Here, we sort the dataset based on the trip distance in descending order.
* Sorting the data helps in identifying extreme values and understanding the distribution of trip distances.

```python
# Sort the data by trip distance from maximum to minimum value
automatidata_sort = automatidata.sort_values(by=['trip_distance'], ascending= False)
automatidata_sort.head(10)
```

The dataset is sorted by the total amount charged for each trip in descending order and prints the top 20 values. 
* This helps in identifying the highest billed trips in the dataset.

```python
# Sort the data by total amount and print the top 20 values
total_amount_sorted = automatidata.sort_values(by=['total_amount'],ascending=False)['total_amount']
total_amount_sorted.head(20)
```

The dataset is sorted by the total amount charged in ascending order and prints the bottom 20 values. 
* This helps in identifying any anomalies or negative amounts in the dataset.

```python
# Sort the data by total amount and print the bottom 20 values
total_amount_sorted.tail(20)
```

This code counts the occurrences of each payment type in the dataset, providing insight into the distribution of payment methods used by passengers.

```python
# How many of each payment type are represented in the data?
automatidata['payment_type'].value_counts()
```

The average tip amount for trips paid for with credit card is calculated and printed. 
* This helps in understanding the tipping behavior of passengers paying with credit card.

Similarly, the average tip amount for trips paid for with cash is calculated and printed. 
* This provides insight into the tipping behavior of passengers paying with cash.

```python
# What is the average tip for trips paid for with credit card?
avg_credit_card_tip = automatidata[automatidata['payment_type']==1]['tip_amount'].mean()
print('average credit card tip :' , avg_credit_card_tip)


# What is the average tip for trips paid for with cash?
avg_cash_tip = automatidata[automatidata['payment_type']==2]['tip_amount'].mean()
print('average cash tip :' , avg_cash_tip)
```

This counts the occurrences of each vendor ID in the dataset, revealing the distribution of trips among different taxi companies.

```python
# How many times is each vendor ID represented in the data?
automatidata['VendorID'].value_counts()
```

Here, the mean total amount charged for trips is calculated for each vendor ID. 
* This provides insight into the average revenue generated by each taxi company.

```python
# What is the mean total amount for each vendor?
automatidata.groupby(['VendorID']).mean(numeric_only=True)[['total_amount']]
```

These lines filter the dataset for credit card payments only and then count the occurrences of each passenger count within the credit card payment subset. 
* This helps in understanding the distribution of passenger counts for credit card payments.

```python
# Filter the data for credit card payments only
credit_card = automatidata[automatidata['payment_type']==1]
# Filter the credit-card-only data for passenger count only
credit_card['passenger_count'].value_counts()
```

 This calculates the average total amount for each passenger count, specifically for trips paid with credit card. 
 * This provides insight into how the passenger count influences the total amount charged for credit card payments.

```python
# Calculate the average tip amount for each passenger count (credit card payments only)
credit_card.groupby(['passenger_count']).mean(numeric_only=True)[['total_amount']]
```

Based on the pre-Explorary analysis of the dataset, the two variables most helpful for building a predictive model for the client are likely to be:

**Trip distance:** This variable appears to have a significant impact on the total amount charged for the taxi trip, as evidenced by its strong correlation with the total amount.

**Payment type:** This variable is crucial as it affects the amount of tip received by the driver, with credit card payments typically resulting in higher tips compared to cash payments.

These variables provide valuable insights into both the distance traveled and the payment method, which are likely to influence the total amount charged and can thus aid in building a predictive model for the client.


### Exploratory data analysis (EDA)

To enhance the exploratory data analysis (EDA) methods, we incorporate more advanced techniques for gaining insights into the dataset.

**Multivariate Analysis:** Explore relationships between multiple variables simultaneously using techniques like pair plots, correlation matrices, and heatmaps.

**Distribution Analysis:** Visualize the distribution of numerical variables and identify any skewed or non-normal distributions. Use techniques like histograms, kernel density estimation (KDE), and box plots.

**Categorical Variable Analysis:** Investigate the distribution and relationships of categorical variables using techniques like bar plots, count plots, and chi-square tests.

**Temporal Analysis:** Analyze trends and patterns over time by aggregating data into different time intervals (e.g., hourly, daily, monthly) and visualizing them using time series plots.


```python
# Feature Engineering
# Create new features like trip duration
automatidata['trip_duration'] = (automatidata['tpep_dropoff_datetime'] - automatidata['tpep_pickup_datetime']).dt.total_seconds() / 60
automatidata.head()
```

```python
# Multivariate Analysis
# Pair plot to visualize relationships between numerical variables
sns.pairplot(automatidata[['trip_distance', 'total_amount', 'trip_duration']])
plt.show()
```

```python
# Correlation matrix and heatmap
corr_matrix = automatidata[['trip_distance', 'total_amount', 'trip_duration']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
